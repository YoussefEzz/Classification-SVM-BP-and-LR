```python
#read the information of A2-ring-merged.txt and load it into a dataframe to plot it
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read A2 ring merged data  
df_merged = pd.read_csv('Data/A2-ring/A2-ring-merged.txt', sep='\t', header=None)
print(df_merged.shape[0])
print(df_merged[df_merged.columns[0]])

#read two input(features) columns
x = df_merged[df_merged.columns[0]]
y = df_merged[df_merged.columns[1]]

#read one output(class) column
output = df_merged[df_merged.columns[2]]
sizes = np.full((1, df_merged.shape[0]), 1)

# Create a scatter plot using Matplotlib
scatter = plt.scatter(x, y, c=output, s=sizes)

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of A2-ring-merged')

# Create a legend using the scatter points and class labels
legend = plt.legend(*scatter.legend_elements(), title='Classes')
# Display the plot
plt.show()
```
