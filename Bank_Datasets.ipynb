```python
# print the unique values in every column
unique_values = {}
for column in df.columns:
    unique_values[column] = df[column].unique()

# Print the unique values for each column
for column, values in unique_values.items():
    print(f"Unique values for column '{column}':")
    print(values)
    print()
```

```python
# drop all rows where at least one column value is unknown 
df_normalized = df
#df_normalized["marital"].value_counts()

# Selecting values based on a condition
condition1 = df_normalized['marital'] == 'unknown'
condition2 = df_normalized['job'] == 'unknown'
condition3 = df_normalized['education'] == 'unknown'
condition4 = df_normalized['default'] == 'unknown'
condition5 = df_normalized['housing'] == 'unknown'
condition6 = df_normalized['loan'] == 'unknown'
selected_values = df_normalized[condition1 | condition2 | condition3 | condition4 | condition5 | condition6]
#print(selected_values.index)

# Dropping the rows with selected values
df_normalized = df_normalized.drop(index = selected_values.index)

print(df_normalized.info())
```

```python
# preprocess
import category_encoders as ce
from sklearn.preprocessing import MinMaxScaler

# encode 10 input categorical features
encoder = ce.OrdinalEncoder(cols=['marital', 'job', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome'])
df_normalized = encoder.fit_transform(df_normalized)
print(df_normalized)
# ... decode 
#df_normalized_reversed = encoder.inverse_transform(df_normalized)

# encode 5 input numerical float features by scaling from -1 to 1
scaler = MinMaxScaler(feature_range=(-1, 1))
# 
columns_to_scale = ['duration', 'emp.var.rate','cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
data_to_scale = df_normalized[columns_to_scale]
scaler.fit(data_to_scale)
scaled_data = scaler.transform(data_to_scale)

rounded_scaled_data = [[round(num, 4) for num in sub_arr] for sub_arr in scaled_data]
df_normalized[columns_to_scale] = rounded_scaled_data

# encode output as 1 for yes and 0 for no
df_normalized["y"].replace({"yes":1 ,"no":0 } , inplace=True)


#print(scaled_data)

# normalize by replacement only
# df_normalized["marital"].replace({"unknown":0 ,"single":1 ,"married":2, "divorced":3} , inplace=True)
# df_normalized["job"].replace({'unknown':0, 'unemployed':1, 'blue-collar':2, 'services':3, 'admin.':4, 'entrepreneur':5, 'self-employed':6
# ,'technician':7, 'management':8, 'student':9, 'retired':10, 'housemaid':11 } , inplace=True)
# df_normalized["education"].replace({'unknown':0, 'illiterate':1, 'basic.4y':2, 'basic.6y':3, 'basic.9y':4, 'high.school':5, 'university.degree':6, 'professional.course':7} , inplace=True)

```
